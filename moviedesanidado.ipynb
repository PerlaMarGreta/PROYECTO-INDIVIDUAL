```python

# Función para comparar dos columnas y verificar si son iguales
def compare_columns(df, col1, col2):
    # Comprobar si las columnas existen
    if col1 in df.columns and col2 in df.columns:
        # Comparar las columnas
        are_equal = df[col1].equals(df[col2])
        return are_equal
    else:
        return None  # Si alguna columna no existe

# Verificar las columnas problemáticas
results = {
    'production_companies vs genres_production_companies': compare_columns(df, 'production_companies', 'genres_production_companies'),
    'production_countries vs genres_production_countries': compare_columns(df, 'production_countries', 'genres_production_countries'),
}

# Imprimir los resultados de la comparación
for comparison, result in results.items():
    if result is None:
        print(f"Una de las columnas en '{comparison}' no existe.")
    else:
        print(f"{comparison}: {'Iguales' if result else 'Diferentes'}")
```

```python
# Eliminar las columnas duplicadas
df = df.drop(columns=['genres_production_companies', 'genres_production_countries'])

# Verificar el resultado
print("Columnas después de eliminar las duplicadas:", df.columns)
```

```python
# Función para encontrar y eliminar columnas duplicadas
def remove_duplicate_columns(df):
    cols_to_remove = []
    for i in range(len(df.columns)):
        for j in range(i + 1, len(df.columns)):
            col1 = df.columns[i]
            col2 = df.columns[j]
            if df[col1].equals(df[col2]):
                print(f"Columnas duplicadas encontradas: {col1} vs {col2}")
                cols_to_remove.append(col2)  # Marcar la segunda columna para eliminar
    # Eliminar las columnas duplicadas
    df = df.drop(columns=cols_to_remove)
    return df

# Aplicar la función al DataFrame
df_cleaned = remove_duplicate_columns(df)

# Verificar el resultado
print("Columnas después de eliminar las duplicadas:", df_cleaned.columns)
```

```python


# Convertir `production_countries` a listas de diccionarios si es una cadena
df['production_countries'] = df['production_countries'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Filtrar filas donde `production_countries` es nulo o vacío
df = df[df['production_countries'].notna() & df['production_countries'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)]

# Expandir la columna y extraer datos
df_expanded = df.explode('production_countries', ignore_index=True)

# Verificar que cada elemento sea un diccionario antes de extraer datos
df_expanded['country_code'] = df_expanded['production_countries'].apply(lambda x: x.get('iso_3166_1') if isinstance(x, dict) else None)
df_expanded['country_name'] = df_expanded['production_countries'].apply(lambda x: x.get('name') if isinstance(x, dict) else None)

# Eliminar la columna original ya que hemos extraído los datos
df_expanded = df_expanded.drop(columns=['production_countries'])

# Verificar el resultado
print(df_expanded[['country_code', 'country_name']].head())

```

CONVERTIR FECHAS

```python
import pandas as pd

# Convertir la columna 'release_date' al formato de fecha AAAA-MM-DD
# coerce: convierte fechas inválidas en NaT, permitiendo detectar errores
df_final['release_date'] = pd.to_datetime(df_final['release_date'], errors='coerce')

# Verificar si alguna fecha no se pudo convertir (NaT indica errores de conversión)
invalid_dates = df_final['release_date'].isnull().sum()
print("Fechas inválidas o no convertidas:", invalid_dates)

# Crear la columna 'release_year' extrayendo el año de 'release_date'
df_final['release_year'] = df_final['release_date'].dt.year

# Verificar el resultado
print(df_final[['release_date', 'release_year']].head())


```

```python
# Convertir 'revenue' y 'budget' a tipo numérico, reemplazando errores de conversión con 0
df_final['revenue'] = pd.to_numeric(df_final['revenue'], errors='coerce').fillna(0)
df_final['budget'] = pd.to_numeric(df_final['budget'], errors='coerce').fillna(0)

# Crear la columna 'return' calculando revenue / budget, manejando divisiones por cero
df_final['return'] = df_final.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] > 0 else 0, axis=1)

# Verificar el resultado
print(df_final[['revenue', 'budget', 'return']].head())
```
